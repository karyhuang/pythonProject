```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests

webpage = requests.get('http://www.simplyrecipes.com/')
page = webpage.text
# print(c)
soup = BeautifulSoup(page, 'html.parser')

#{'class':'ec_statements'}
x=soup.find_all("div", class_="grd-tile-box") 
# print(x) will list all the div tags as a list
for item in x:
    print(item)
    print('--------------------------------------------------')
```

```python
# print out all the recipe names
for item in x:
    print(item.h2.get_text())
```

```python
recipe = {} # create an empty dictionary called recipe
for item in x:
    recipe[item.h2.get_text()] = {} # using recipe names to create a bunch of empty dictionaries within the recipe dictionary
# here we put recipe names as keys, and each with empty value

for key,value in recipe.items(): 
    print(key,value)
```

```python
# populate the recipe dictionaries with content, starting with the link
for item in x:
    recipe[item.h2.get_text()]["link"] = item.a['href']
for key,value in recipe.items(): 
    print(key,value, sep='\n')
    print('--------------------------------------------------')
```

```python
x[0].find(itemprop="author")
```

```python
# add the author
for item in x:
    author = item.find(itemprop="author").get_text()
    recipe[item.h2.get_text()]["author"] = author
    
for key,value in recipe.items(): 
    print(key,value, sep='\n')
    print('--------------------------------------------------')
```

```python
# the second recipe does not have a cooking time
x[1].find(itemprop="totalTime").get_text()
```

```python
# cannot add the cooking time because not all recipes have cooking time
for item in x:
    time = item.find(itemprop="totalTime").get_text()
    recipe[item.h2.get_text()]["cooking-time"] = time
    
for key,value in recipe.items(): 
    print(key,value, sep='\n')
    print('--------------------------------------------------')
```

```python
for item in recipe.keys():
    print (item + ": " + "\n\t" + "link: " + recipe[item]["link"] + "\n\t" + "author: " + recipe[item]["author"] + "\n\n") 
```
